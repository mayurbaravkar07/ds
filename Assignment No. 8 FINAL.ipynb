```python
import pandas as pd
```

```python
df=pd.read_csv("Titanic.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['fare'])
```

```python
plt.scatter(df['age'],df['fare'])
```

```python
sns.distplot(df['fare'])
```

```python
sns.boxplot(df['fare'])
```

```python
sns.jointplot(x='age', y='fare', data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.rugplot(df['fare'])
```

```python
sns.barplot(x='sex', y='age', data=df)
```

```python
sns.countplot(x='sex', data=df)
```

```python
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x='sex', y='age', data=df, hue="survived")
```

```python
sns.violinplot(x='sex', y='age', data=df)
```

```python
sns.stripplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df)
```

```python
sns.swarmplot(x='sex', y='age', data=df, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data=df, hue='survived', split=True)
```

```python
sns.catplot(x='sex', y='age', data=df, hue='survived', split=True)
```

```python
sns.dogplot(x='sex', y='age', data=df, hue='survived', split=True)
```

```python
sns.pointplot(x='sex', y='age', data=df, hue='survived', split=True)
```
